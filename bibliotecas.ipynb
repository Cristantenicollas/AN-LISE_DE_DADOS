```python
#Importação de bibliotecas 
import numpy as np #np será o apelido
```

```python
temperaturas = np.array([30,40,50,60.70,80,90,100])
print(temperaturas)
print(type(temperaturas))
```

```python
# Média aritmédica
media = np.mean(temperaturas)
print(f"{media: .2f}")
```

```python
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
#Sálarios x 1000
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```

Desafio final:
- Criar um array de habitantes de 5 cidades do estado de sp.
- Maior população
- Menor população
- Média de população
- mediana

```python
população = np.array([11904961, 1349100, 1187974, 841154, 782448])
maior_população = np.max(população)
menor_população = np.min(população)
media_população = np.mean(população)
mediana_população = np.median(população)
print(f"A maior população é {maior_população}")
print(f"A menor população é {menor_população}")
print(f"A media da população é {media_população}")
print(f"A mediana da população é {mediana_população}")
```

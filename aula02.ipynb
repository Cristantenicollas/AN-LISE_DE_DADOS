Programa que avalia as notas que o usuário do nosso aplicativo inseriu.

```python
avaliacao = int(input("Digite a sua avaliação: 0 - 10 estrelas"))

if avaliacao > 8:
    print("Nota exelente!")
elif avaliacao > 6:
    print("Avaliação regular!")
elif avaliacao > 4:
    print("Avaliação ruim!")
else:
    print("Avaliação péssima!")
```

Sistema que ira verificar se o filme avaliado é bom ou não.
- Criar uma variavel que irá armazenar o nome do filme
- Criar uma variavel que irá armazenar a nota do filme (0 - 5)
- Condição: se possuir 5 estrelas, filme vencedor do Oscar.
- Se possuir de 3 a 4,99, filme bom para assistir
- abaixo disso, filme não indicado

```python
Filme = input("Digite o nome do filme: ")
nota_do_filme = int(input("Digite a nota do filme de 0-5 para o filme: "))

if nota_do_filme == 5:
    print(f"O Filme {Filme} é o Vencedor Do Oscar!!!")
elif nota_do_filme > 3 :
    print(f"O Filme {Filme} é bom para assistir!!")
else:
    print(f"O Filme {Filme} não indicado!")
```

```python
Filme = input("Digite o nome do filme: ")
nota_do_filme = int(input("Digite a nota do filme de 0-5 para o filme: "))
if nota_do_filme <= 0 or nota_do_filme > 5 :
    print("❌ NOTA INVALIDA!!")
else:
    if nota_do_filme == 5:
        print(f"O Filme {Filme} é o Vencedor Do Oscar!!!")
    elif nota_do_filme > 3 :
        print(f"O Filme {Filme} é bom para assistir!!")
    else:
        print(f"O Filme {Filme} não indicado!")
```

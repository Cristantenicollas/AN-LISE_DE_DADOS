N-gramas (n-gramas)

```python
frase = "Ele sabia que eu sabia trabalhar com IA e sabia que eu estava estudando tecnologia sabia disso"
```

```python
# O split
palavras = frase.split() 
palavras
```

```python
import pandas as pd
tabela = pd.DataFrame({"palavras":palavras})
tabela
```

```python
tabela["proxima_palavra"] = tabela["palavras"].shift(-1)
tabela
```

```python
filtro = tabela["palavras"] == "que"
tabela[filtro]
```

```python
tabela[filtro]["palavras"].value_counts()
```

```python
tabela[filtro]["proxima_palavra"].value_counts()
```

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize=True)
```

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```

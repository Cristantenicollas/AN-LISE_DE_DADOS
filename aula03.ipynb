````python
import pandas as pd 
import numpy as np
````

````python
quantidade_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nomes_produtos = ["Pastel","Caçulinha","Refrigerante","Combos"]

array_produtos = np.array(quantidade_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
for nome,preco in zip(nomes_produtos,faturamento):
    print(f"{nome} : {preco}")
````

### Exercício de fixação 
- Criar uma lista que ira possuir 10 nomes de produtos.
- Criar uma lista que ira possuir o preço de custo (Atacado).
- Criar uma lista que irá possuir o preço de venda (Varejo).
- Ao final, exibir o lucro obtido em cada produto.

````python
quantidade_produtos = [10,6,10,30,12,6,24,12,24,12]
precos_produtos_atacado = [20,14,55,4,5,30,18,30,20,40]
precos_produtos_varejo = [5,35,10,8,10,7,4,7,4,10]
nomes_produtos = ["Kit de Meias","Pulverizador Manual 2L","Toalha Umedecida","Esmaltes Variados","Porta Comprimidos Semanal","Bebidas Fardo Refri 2L","Detergente Líquido 500ml","Creme Dental","Sabonete em Barra","Desodorante Spray/Roll-on"]

array_precos01 = np.array(precos_produtos_atacado)
array_precos02 = np.array(precos_produtos_varejo)

lucro = array_precos01 - array_precos02
lucro_obtido = np.sum(lucro)
print(f"O lucro obtido foi R${lucro_obtido}")

for nome,preco in zip(nomes_produtos,lucro):
    print(f"{nome} : {preco}")
````
